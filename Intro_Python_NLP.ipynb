```python
#views del fin de semana usando TQDM
```

```python
def weekendindentifier(day):
  if day in ['Saturday','Sunday']:
    return True
  else:
    return False
```

```python
weekendindentifier('Monday')
```

```python
from tqdm import tqdm
tqdm.pandas()
```

```python
df['is_weekend'] = df['Dia_de_la_semana'].progress_apply(weekendindentifier)
```

```python
mapeador = {'Monday':'Lunes','Tuesday':'Martes','Wednesday':'Miercoles','Thursday':'Jueves','Friday':'Viernes','Saturday':'Sabado','Sunday':'Domingo'}
```

```python
df['dia_esp']=df['Dia_de_la_semana'].map(mapeador)
df.head()
```

```python
#del canal de dross decir cual fue el video más comentado por mes. El mes debe estar en español
```

```python
#organizar la traduccion de meses
```

```python
mapeador2={'January':'Enero','February':'Febrero','March':'Marzo','April':'Abril','May':'Mayo','June':'Junio','July':'Julio','August':'Agosto','September':'Septiembre','October':'Octubre','November':'Noviembre','December':'Diciembre'}
```

```python
df['Mes'] = df['publish_time'].dt.month_name()
df.head()
```

```python
df['mes_esp']=df['Mes'].map(mapeador2)
df.head()
```

```python
dross_df=df[df['channel_title'].str.contains('Dross',case=False)]
dross_df
```

```python
#del canal de dross decir cual fue el video más comentado por mes.
```

```python
meses=dross_df['mes_esp'].unique()
meses
```

```python
for mes in meses:
  filtado=dross_df[dross_df['mes_esp']==mes]
  resultado=filtado.sort_values(by='comment_count',ascending=False).head(1)['title']
  print(f"{mes}:{resultado}")
```
