```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/LIDCI/5/PROCESAMIENTO DE LENGUAJE NATURAL/realdonaldtrump.csv')
```

```python
df.head()
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df['year']=df['date'].dt.year
```

```python
df.head()
```

```python
df = df[df['year']==2020]
df.head()
```

```python
tweets=''.join(df['content'])
df.head()
```

```python
import re
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
stopwords_en=stopwords.words('english')
stopwords_en.append('twitter')
```

```python
stopwords_en
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
tweets = tweets.lower()
tweets = re.sub(r'@[a-z0-9]+',' ',tweets)
tweets = re.sub(r'\w+://\s+',' ',tweets)
tweets = re.sub(r'\w+://\s?',' ',tweets)
tweets = re.sub(r'[^0-9a-z]',' ',tweets)
tweets
```

```python
tweets_tokenzados = word_tokenize(tweets)
tweets_tokenzados
```

```python
test = [numero for numero in range(10)]
test
```

```python
tweets_non_stop=[token for token in tweets_tokenzados if token not in stopwords_en]
```

```python
final = [lemmatizer.lemmatize(word) for word in tweets_non_stop]
```

```python
final = ' '.join(final)
final
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud = WordCloud(width=1200, height=1200, colormap='YlOrBr_r', background_color='black', min_font_size=6).generate(final)

plt.figure(figsize=(5,5))
plt.imshow(wordcloud)
plt.axis('off')
plt.tight_layout()
plt.show()
```
