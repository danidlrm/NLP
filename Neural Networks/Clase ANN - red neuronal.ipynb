```python
import re
import nltk
import numpy as np
import pandas as pd
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
import contractions

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('punkt_tab')
stop_words_en = stopwords.words('english')
lemmatizer = WordNetLemmatizer()
```

```python
headlines=''.join(df['headline'])
df.head()
```

```python
def vectorize(texto):
    texto=texto.lower()
    texto = contractions.fix(texto)
    texto = re.sub(r'[^a-zA-Z\s]', '', texto)
    texto_tokenized= word_tokenize(texto)
    texto_no_stop = [token for token in texto_tokenized if token not in stop_words_en]
    final= [lemmatizer.lemmatize (word) for word in texto_no_stop]
    texto_lemmatized = [lemmatizer.lemmatize(word) for word in texto_no_stop]
    final = ' '.join(final)
    vector = vectorizer.transform([final])
    vector = vector.reshape(1,-1[0])

    return vector
```

```python
vectorize('this is a test')
```

```python
from tqdm import tqdm
nltk.download('punkt_tab')
tqdm.pandas()
df['vector'] = df['headline'].progress_apply(vectorize)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
X = df['vector']
X = np.concatenate(X,axis=0).reshape(1,-300)
y = df['clickbait']

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    train_size = 0.7,
                                                    random_state = 101)
```
